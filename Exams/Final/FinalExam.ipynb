As before, write a function `random_walk(num_steps,init_position)` that takes an integer for `num_steps` corresponding to how many steps the random walker will take, and a numpy array for `init_position` with x and y coordinates of the starting point. The function should return the full sequence of positions of the walker as they take `num_steps` steps
1. To determine the next location of the walker at a given step, the walker uniformly randomly chooses to walk one of 4 possible directions: up, down, left, or right and takes a step of length 1 in either direction (e.g., if the walker starts at (0,0), their possible moves for the first step are up=(0,1), down=(0,-1), left=(-1,0), right=(1,0))
1. Call your `random_walk` function 5 times with initial position of (0,0) and 10,000 steps and plot the random walk for each function call using pyplot (here is an example of what it should look like; if the image does not render in GitHub, open 2drandomwalk.png in the Final folder)

1. Call your modified random walk function 5 times and plot the results.
1. What do you notice? Is there a difference in behavior compared to the original version?

```python
# Write your code here
import matplotlib.pyplot as plt

switch = {
    1: (0,1),  # Up
    2: (0,-1), # Down
    3: (-1,0), # Left
    4: (1,0)  # Right
}
def random_walk(num_steps, init_position, probability):
    steps = []
    steps.append(init_position)
    for x in range(0, int(num_steps)):
        if probability == 0:
            random_value = np.random.randint(1, 5)
            list2 = switch.get(random_value)
        else: 
            random_value = np.random.choice([1,2,3,4], 1, p=[0.7, 0.1, 0.1, 0.1])
            list2 = switch.get(random_value[0])
        # Random Walk
        if steps[x-1] != None:
            list1 = steps[x-1]
            zipped_lists = zip(list1, list2)
            sum = [x + y for (x, y) in zipped_lists]
            steps.append(sum)
        else:
            steps.append(switch.get(random_value))
#     print(steps)
    return(steps)
    
x = []
y = []
data = random_walk(10000, (2,1), 0)

for i in data:
    x.append(i[0])
    y.append(i[1])
    
# print(data)
# print(x)
# print(y)
def plotData(x, y):
    plt.plot(x, y)
    plt.show()

plotData(x, y)

#####################
# What do you notice?
#####################
print('#####################\n# What do you notice?\n#####################')
print('# I notice that each time I render the result, due to the result and walk being done in random position,')
print('## it can be said that it will give different results each time')


# Modify your random walk such that the walker is more likely to travel up (specifically, they will step up with probability .7 and the remaining directions each with probability .1).
x = []
y = []
data = random_walk(10000, (2,1), 1)
for i in data:
    x.append(i[0])
    y.append(i[1])


# Call your modified random walk function 5 times and plot the results.
plotData(x, y)
plotData(x, y)
plotData(x, y)
plotData(x, y)
plotData(x, y)

# What do you notice? Is there a difference in behavior compared to the original version?
print("# What do you notice? Is there a difference in behavior compared to the original version?")
print("The difference is huge in original and modified behavior because in the modified version, " + 
      "most values are looking forward to get the `Up` value because it has probability of 0.7 " + 
      "while others have 0.1 probability. \n\n" + 
      "Such that, due to high probabilty in one caused the increase its obtainable chances, thus all made a nearly identical paterren "+ 
      "excluding the original who kept all probabilities equal.")
```
